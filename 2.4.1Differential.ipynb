# 2.4 微分計算

## 2.4.1 １変数関数の微分

```julia
# using Pkg
# Pkg.add("PyPlot")
```

```julia
using PyPlot
```

```julia
# f(x)を２次関数として定義
f(x) = -(x + 1)*(x - 1)
```

```julia
# hを微小な値として設定(10のマイナス10乗)
h = 1.0e-10
```

```julia
# 導関数f'の近似式
fʼ(a) = (f(a+h) - f(a)) / h
```

```julia
# 関数の可視化範囲
xs = range(-1, 1, length=100)

fig, axes = subplots(2, 1, figsize=(4,6))

# 関数のプロット
axes[1].plot(xs, f.(xs), "b")
axes[1].grid()
axes[1].set_xlabel("x"), axes[1].set_ylabel("y")
axes[1].set_title("funcion f(x)")

# 導関数のプロット
axes[2].plot(xs, fʼ.(xs), "r")
axes[2].grid()
axes[2].set_xlabel("x"), axes[2].set_ylabel("y")
axes[2].set_title("derivative fʼ(x)")

tight_layout()
```
