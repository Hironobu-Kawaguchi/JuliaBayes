```julia
# 1つ目のデータ
Y1 = [1, 7, 11, 13, 15, 16]

# 2つ目のデータ
Y2 = [15, 3, 13, 2, 7, 1]
```

```julia
# グラフを作成
fig, ax = subplots()

# 折れ線グラフを描画
ax.plot(Y1)
ax.plot(Y2)

# 軸の名称を設定
ax.set_xlabel("index")
ax.set_ylabel("Y")

# グラフのタイトルを設定
ax.set_title("my graph")
```

```julia
fig, axes = subplots(1, 2, figsize=(10, 3))
axes[1].plot(Y1)
axes[2].plot(Y2)
```

```julia
# 2次関数の定義
f(x) = x^2

# -3から3まで、10個の等間隔の点列を生成
xs = range(-3, 3, length=10)

fig, ax = subplots()

# 関数の描画
ax.plot(xs, f.(xs), "-")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("y=x^2")
```

```julia
# 2次関数の定義
f(x) = x^2

# -3から3まで、100個の等間隔の点列を生成
xs = range(-3, 3, length=100)

fig, ax = subplots()

# 関数の描画
ax.plot(xs, f.(xs), "-")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("y=x^2")
```

```julia
r = 1.0
fx(θ) = r*(θ - sin(θ))
fy(θ) = r*(1 - cos(θ))
θs = range(-3pi, 3pi, length=100)
fig, ax = subplots()
ax.plot(fx.(θs), fy.(θs))
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("cycloid")
```

```julia
# シグモイド関数を定義
sig(x) = 1/(1 + exp(-x))
xs = range(-5, 5, length=100)
fig, ax = subplots()
ax.plot(xs, sig.(xs))
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("sigmoid")
```

```julia
# 可視化に使う乱数を生成
X = randn(1000)
Y = rand(1000)

fig, ax = subplots()

# ax.plot(X, Y, "o")などでも可
ax.scatter(X, Y)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("cycloid")
```

```julia
X = randn(1000)
fig, ax = subplots()
ax.hist(X)
```

```julia
fig, ax = subplots()
ax.hist(X, bins=50)
```

```julia
# 2変数関数を定義
fz(x, y) = exp(-(2x^2+y^2+x*y))

# x軸とy軸の可視化範囲を定義
xs = range(-1, 1, length=100)
ys = range(-2, 2, length=100)

fig, axes = subplots(1, 2, figsize=(8,4))

# 1つ目のsubplotには数値をつける
cs = axes[1].contour(xs, ys, fz.(xs', ys))
axes[1].clabel(cs, inline=true)

# 2つ目のsubplotにはカラーバーをつける
cs = axes[2].contourf(xs, ys, fz.(xs', ys))
fig.colorbar(cs)

# subplots間の余白の大きさを自動調整
tight_layout()
```
