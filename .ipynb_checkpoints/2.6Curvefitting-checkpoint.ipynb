# 2.6　最適化によるカーブフィッティング

```julia
using PyPlot
using Statistics
using ForwardDiff
```

```julia
# 多変数解析のための勾配法
function gradient_method(f, x_init, η, maxiter)
    # 最適化過程のパラメータを格納する配列
    x_seq = Array{typeof(x_init[1]), 2}(undef, length(x_init), maxiter)
    
    # 勾配
    ∇f(x) = ForwardDiff.gradient(f, x)
    
    # 初期値
    x_seq[:, 1] = x_init
    
    # メインの最適化ループ
    for i in 2:maxiter
        x_seq[:, i] = x_seq[:, i-1] + η*∇f(x_seq[:, i-1])
    end
    
    x_seq
end
```

```julia
# 学習用の入力値集合
X_obs = [1, 2, 4]

# 学習用の出力値集合
Y_obs = [10, 35, 72]

# データの可視化
fig, ax = subplots()
ax.scatter(X_obs, Y_obs)
ax.set_xlabel("x"), ax.set_ylabel("y")
ax.set_title("data")
ax.grid()
```

```julia
# 適当な傾きパラメータおよび切片パラメータを設定
w = [12.0, 10.0]

# 予測に使う関数
f(x) = w[1] * x + w[2]

# 関数を可視化する範囲
xs = range(0, 5, length=100)

# データと予測関数の可視化
fig, ax = subplots()
ax.plot(xs, f.(xs), label="function")
ax.scatter(X_obs, Y_obs, label="data")
ax.set_xlabel("x"), ax.set_ylabel("y")
ax.grid()
ax.legend()
```

```julia
# 誤差関数の定義
E(w) = sum([(Y_obs[n] - (w[1] * X_obs[n] + w[2]))^2 for n in 1:length(X_obs)])
```

```julia
# 最適化するパラメータの初期値
w_init = [0.0, 0.0]

# 最適化計算の回数
maxiter = 500

# 学習率
η = 0.01

# 最適化の実行
F(w) = -E(w)
w_seq = gradient_method(F, w_init, η, maxiter)
f_seq = [F(w_seq[:,i]) for i in 1:maxiter]
```

```julia
w₁, w₂ = w_seq[:, end]
println("w₁ = $(w₁), w₂ = $(w₂)")
```

```julia
# 目的関数の値をステップごとにプロット
fig, ax = subplots(figsize=(8,4))
ax.plot(f_seq)
ax.set_xlabel("iteration"), ax.set_ylabel("f")
ax.grid()
```

```julia
# 予測に使う関数
f(x) = w₁ * x + w₂

# 関数を可視化する範囲
xs = range(0, 5, length=100)

# データと予測関数の可視化
fig, ax = subplots()
ax.plot(xs, f.(xs), label="f(x)")
ax.scatter(X_obs, Y_obs, label="data")
ax.set_xlabel("x"), ax.set_ylabel("y")
ax.grid()
ax.legend()
```

```julia
function linear_fit(Y, X)
    N = length(Y)
    w₁ = sum((Y .- mean(Y)) .* X) / sum((X .- mean(X)) .* X)
    w₂ = mean(Y) - w₁ * mean(X)
    w₁, w₂
end
```

```julia
w₁, w₂ = linear_fit(Y_obs, X_obs)
println("w₁ = $(w₁), w₂ = $(w₂)")
```
