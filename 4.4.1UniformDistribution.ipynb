# 4.4 連続型確率分布

```julia
using Distributions
using PyPlot
```

```julia
function set_options(ax, xlabel, ylabel, title;
                     grid=true, gridy=false, legend=false)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    if grid
        if gridy
            ax.grid(axis="y")
        else
            ax.grid()
        end
    end
    legend && ax.legend()
end
```

## 4.4.1 一様分布

```julia
# パラメータ
a = 0
b = 1

# 一様分布の作成
d = Uniform(a, b)
```

```julia
# サンプリング
X = rand(d, 10000)
X'
```

```julia
fig, ax = subplots()
ax.hist(X)
set_options(ax, "x", "frequency", "histogram"; gridy=true)
```
